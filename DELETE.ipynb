```python
if fng_str:
    nuclides = ['nb93', 'al27', 'ni58', 'au197']

    old_file = 'experiment.h5'
    new_file = 'experiment_1.h5'

    for i,nu in enumerate(nuclides):
        with h5py.File(old_file) as f:
            rr_onaxis1 = pd.DataFrame(f[f'reaction_rates_onaxis/group_1/{nu}/table'][()]).drop(['index'], axis=1).rename(columns={'y(cm)':'Shield depth (cm)', 'mean':'mean', 'std. dev.':'std.dev.'})
            rr_offaxis = pd.DataFrame(f[f'reaction_rates_offaxis/{nu}/table'][()]).drop(['index'], axis=1).rename(columns={'y(cm)':'Shield depth (cm)', 'mean':'mean', 'std. dev.':'std.dev.'})
            rr_onaxis2 = pd.DataFrame(f[f'reaction_rates_onaxis/group_2/{nu}/table'][()]).drop(['index'], axis=1).rename(columns={'y(cm)':'Shield depth (cm)', 'mean':'mean', 'std. dev.':'std.dev.'})

        rr_onaxis1 = rr_onaxis2.astype({'Shield depth (cm)':str})    
        rr_onaxis2 = rr_onaxis2.astype({'Shield depth (cm)':str})
        rr_offaxis = rr_onaxis2.astype({'Shield depth (cm)':str})

        rr_onaxis1.to_hdf(new_file, f'rr_onaxis1_{nu}', mode='a',format='table',data_columns=True, index=False)
        rr_onaxis2.to_hdf(new_file, f'rr_onaxis2_{nu}', mode='a',format='table',data_columns=True, index=False)
        rr_offaxis.to_hdf(new_file, f'rr_offaxis_{nu}', mode='a',format='table',data_columns=True, index=False)

        with h5py.File(new_file, 'a') as f:
            f[f'rr_onaxis1_{nu}/table'].attrs['x_axis'] = 'Shield depth (cm)'

    with h5py.File(old_file) as f:
        heating = pd.DataFrame(f[f'heating/results/table'][()]).drop(['index'], axis=1).rename(columns={'y(cm)':'Shield depth (cm)', 'mean':'mean', 'std. dev.':'std.dev.'})

    heating = heating.astype({'Shield depth (cm)':str})   
    heating.to_hdf(new_file, f'nuclear_heating', mode='a',format='table',data_columns=True, index=False)

    with h5py.File(new_file, 'a') as f:
        f['nuclear_heating/table'].attrs['x_axis'] = 'Shield depth (cm)'
        f.attrs['literature'] = ['https://doi.org/10.1016/S0920-3796(00)00205-2']
        f.attrs['where'] = ['ENEA, Frascati (Italy)']
        f.attrs['when'] = ['1997-1998']
```
