```python
import h5py
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import openmc
from helpers import *
```

```python
# read sinbad data
sinbad_file = h5py.File('sinbad_data.h5', 'r')
```

```python
path = 'reaction_rates_onaxis/group_1/'

# initialization
foil_list = ['nb93', 'al27', 'ni58', 'au197']
reaction_labels = ['$^{93}$Nb(n,2n)$^{92m}$Nb', '$^{27}$Al(n,\u03B1)$^{24}$Na', '$^{58}$Ni(n,p)$^{58}$Co', '$^{197}$Au(n,\u03B3)$^{198}$Au']
ylabel = r'Reaction Rate $(\frac{1}{src \cdot atom})$'

for i,foil in enumerate(foil_list):
    measured = sinbad_file[path + f'measured/{foil}/table']
    mcnp_eff3 = sinbad_file[path + f'mcnp-eff3/{foil}/table']
    mcnp_fendl1 = sinbad_file[path + f'mcnp-fendl1/{foil}/table']
    try:
        mcnp_fendl2 = sinbad_file[path + f'mcnp-fendl2/{foil}/table']
    except:
        continue
    
    dtype_label = reaction_labels[i]


    my_xticks = np.arange(len(measured))

    foils_onaxis_group1 = VisualizeResults()
    foils_onaxis_group1.add_measured_data(measured, my_xticks, ylabel, dtype_label)
    foils_onaxis_group1.add_computed_data(mcnp_eff3, my_xticks, 'd', 'tab:green', 'mcnp-eff3')
    foils_onaxis_group1.add_computed_data(mcnp_fendl1, my_xticks, 'o', 'c', 'mcnp-fendl1')
    foils_onaxis_group1.add_computed_data(mcnp_fendl2, my_xticks, '^', 'tab:pink', 'mcnp-fendl2')
```

```python
path = 'reaction_rates_onaxis/group_2/'

# initialization
foil_list = ['nb93', 'al27', 'ni58', 'au197']
reaction_labels = ['$^{93}$Nb(n,2n)$^{92m}$Nb', '$^{27}$Al(n,\u03B1)$^{24}$Na', '$^{58}$Ni(n,p)$^{58}$Co', '$^{197}$Au(n,\u03B3)$^{198}$Au']
ylabel = r'Reaction Rate $(\frac{1}{src \cdot atom})$'

for i,foil in enumerate(foil_list):
    measured = sinbad_file[path + f'measured/{foil}/table']
    mcnp_eff3 = sinbad_file[path + f'mcnp-eff3/{foil}/table']
    mcnp_fendl1 = sinbad_file[path + f'mcnp-fendl1/{foil}/table']
    try:
        mcnp_fendl2 = sinbad_file[path + f'mcnp-fendl2/{foil}/table']
    except:
        continue
    
    dtype_label = reaction_labels[i]


    my_xticks = np.arange(len(measured))

    fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(6, 5), gridspec_kw={'height_ratios': [2, 1.25]}, constrained_layout=True)

    foils_onaxis_group2 = VisualizeResults(fig, ax1, ax2, my_xticks)
    foils_onaxis_group2.add_measured_data(measured, ylabel, dtype_label)
    foils_onaxis_group2.add_computed_data(mcnp_eff3, 'd', 'tab:green', 'mcnp-eff3')
    foils_onaxis_group2.add_computed_data(mcnp_fendl1, 'o', 'c', 'mcnp-fendl1')
    foils_onaxis_group2.add_computed_data(mcnp_fendl2, '^', 'tab:pink', 'mcnp-fendl2')
```
